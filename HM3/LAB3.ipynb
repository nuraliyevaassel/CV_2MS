```python
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
from skimage import feature
import random
```

```python
image = data.camera()
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        image[i,j] += random.randint(-20,20)
```

```python
plt.imshow(image, cmap = plt.cm.gray)
```

```python
mm = np.median(image)
print(mm)
binary = image > mm
plt.imshow(binary, cmap = plt.cm.gray)
```

```python
edges1 = feature.canny(image, sigma=3.5, high_threshold=45, low_threshold=30)
plt.imshow(edges1, cmap=plt.cm.gray)
```

```python
original = data.camera()
img_edges = feature.canny(original,sigma=2,)
plt.imshow(img_edges, cmap = plt.cm.gray)
```
